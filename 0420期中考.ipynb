```
test <- data.frame(gp = factor(rep(letters[1:3], each = 20)), y = rnorm(30))
ggplot(test, aes(x = gp, y = y)) + geom_point()
```

```
ggplot(test, aes(x = gp, y=y)) + geom_boxplot()
```

```
test$z <- test$y+rnorm(30)
ggplot(test, aes(x = z, y = y)) + geom_point() + facet_grid(gp~.)
```

```
ggplot(test, aes(x = z, y = y)) + geom_point() + facet_grid(gp~.) + geom_smooth()
```

```
ggplot(test, aes(x = z, y = y, color = gp)) + geom_line()
```
