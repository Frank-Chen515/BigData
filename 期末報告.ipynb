```
pkgs <- c("magrittr", "dplyr")
install.packages(pkgs)
library(magrittr)
library(dplyr)
```

```
library(readr)
cancer<-read.csv("data_cancer.csv",fileEncoding = "Big5")
```

```
cancer %>%
  dplyr::filter(縣市別 == "台閩地區",癌症別 == "全癌症",性別 == "全") %>%
  dplyr::select(癌症診斷年,癌症發生數) %>%
  arrange(desc(癌症發生數)) #先做每年的總癌症發生數
```

```
info<-cancer %>%
  dplyr::filter(縣市別 == "台閩地區",癌症別 == "全癌症",性別 == "全") %>%
  dplyr::select(癌症診斷年,癌症發生數) 
qplot(癌症診斷年,癌症發生數,data=info,geom = c("point", "smooth")) #先做每年的總癌症發生數(圖形)
```

```
info<-cancer %>%
  dplyr::filter(縣市別 == "台閩地區",癌症別 == "全癌症") %>%
  dplyr::select(癌症診斷年,癌症發生數,性別)
qplot(癌症診斷年,癌症發生數,data=info,color = 性別) #每年的總癌症發生數(性別圖形)
```

```
cancer %>%
  dplyr::filter(癌症診斷年 == "2018",癌症別 == "肺、支氣管及氣管",縣市別 != "台閩地區",性別 == "全") %>%
  dplyr::select(癌症診斷年,縣市別,癌症發生數,性別) %>%
  arrange(desc(癌症發生數)) #2018檢測肺癌發生數
```

```
info <- cancer %>%
  dplyr::filter(癌症別 == "肺、支氣管及氣管",縣市別 == "台中市") %>%
  dplyr::select(癌症診斷年,癌症發生數,性別)
qplot(癌症診斷年,癌症發生數,data=info,color = 性別) #查看台中肺、支氣管及氣管癌的歷年分析圖
```

```
info <- cancer %>%
  dplyr::filter(癌症別 == "肺、支氣管及氣管",縣市別 == "台中市",性別=="全") %>%
  dplyr::select(粗率,癌症診斷年)
qplot(癌症診斷年,粗率,data=info,geom = c("point", "path")) #查看台中肺、支氣管及氣管癌的歷年粗率分析
```

```
info<-cancer %>%
  dplyr::filter(縣市別 == "台閩地區",癌症別 == "子宮頸",性別 == "女") %>%
  dplyr::select(癌症診斷年,粗率,平均年齡) 
qplot(癌症診斷年,粗率,data=info,geom = "line") #查看全台灣女性子宮頸癌之粗率(每十萬人口)
qplot(癌症診斷年,平均年齡,data=info) #查看全台灣女性子宮頸癌之平均年齡
```
